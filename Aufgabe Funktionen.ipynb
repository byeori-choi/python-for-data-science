Schreibe eine Funktion `z`, die die Vielfachen einer Zahl (bis 10x Zahl) ausgibt. Beispiel: z(6) soll ausgeben: 

```

### Aufgabe 2: 

Erstelle eine Funktion, die eine Temperatur von Celsius in Fahrenheit umrechnen kann.

Du kannst hierfür folgende Formel verwenden: °F = °C * 1,8 + 32

Gewünschte Ausgabe: 212 (oder 212.0)

````python
def celsius_to_fahrenheit(celsius):
    # Passe diese Funktion an
    fahrenheit = celsius * 1.8 + 32
    return fahrenheit

print(celsius_to_fahrenheit(100))
````

### Aufgabe 3: Häufigkeit von Wörtern ermitteln

Schreibe eine Funktion `count_words(text, pattern)`, die folgende Parameter übergeben bekommt:

- text: Das ist der gesamte Text
- pattern: Hiernach soll gesucht werden

Diese Funktion soll dann die Häufigkeit angeben, wie oft `pattern` in String `text` enthalten ist. Wichtig: Hierbei soll Groß- und Kleinschreibung ignoriert werden, du kannst hierfür die `.lower()`-Funktion verwenden.

Öffne anschließend den text von Faust, und ermittle, wie oft das Wort "Faust", "Gretchen" und "Mephisto" im Text vorkommt.

**Hinweis:** Die Lizenzhinweise mit am Anfang des eBooks zählen wir jetzt mal mit zum Buch.

**Hinweis:** Mit `.count('abc')` wird ausgegeben, wie oft der Teilstring `abc` in einem String vorkommt. Wichtig: Diese Funktion unterscheidet zwischen Groß- und Kleinschreibung:

````python
text = "abcdef abcdef abcdef"
print(text.count("abc"))
````

Die Ausgabe soll so aussehen:

    278-mal kommt Faust vor.
    39-mal kommt Gretchen vor.
    284-mal kommt Mephisto vor.

````python
def count_words (text,pattern):
    haeufigkeit = text.count(pattern)
    return haeufigkeit

with open("faust.txt", encoding = "utf-8") as file:
    contents = file.read().lower()
    print(str(count_words(contents,"faust")) + "-mal kommt Faust vor.")
    print(str(count_words(contents,"gretchen")) + "-mal kommt Gretchen vor.")
    print(str(count_words(contents,"mephisto")) + "-mal kommt Mephisto vor.")
    
````

````python
def count_words (text,pattern):
    return text.lower().count(pattern.lower())

with open("faust.txt", encoding = "utf-8") as file:
    contents = file.read()

    print(str(count_words(contents,"Faust")) + "-mal kommt Faust vor.")
    print(str(count_words(contents,"Gretchen")) + "-mal kommt Gretchen vor.")
    print(str(count_words(contents,"Mephisto")) + "-mal kommt Mephisto vor.")

    words = ["Faust", "Gretchen", "Mephisto"]
    for word in words:
        print(str(count_words(contents, word)) + "-mal kommt " + word + " vor.")
````

````python
all_characters = "abcdefghijklmnopqrstuvwxyzüöäß" # alternativ: strings.ascii_letters

def clean_word(word):
    w = ""
    for character in word.lower():
        if character in all_characters:
            w = w + character
    return w

with open("faust.txt", encoding="utf-8") as file:
    contents = file.read()
    # Schreibe hier deinen Code
    words = [clean_word(word) for word in contents.split()]
    word_count = {}
    for word in words:
        if word in word_count:
            word_count[word] = word_count[word] + 1 # Danach wird das word im Dictionary betrachtet, und die Value nimmt um 1 zu.
        else:
            word_count[word] = 1 # Am Anfang ist word nicht im Dictionary word_count. mit else:-Block wird das word hinzugefügt, und die Value ist 1(-mal).

    max_value = 0
    for key, value in word_count.items():
        if value > max_value:
            max_value = value
            max_key = key
            
    print(word_count)
    print("Am Häufigsten kam vor: " + max_key + " (" + str(max_value) + "-mal)" )
````
